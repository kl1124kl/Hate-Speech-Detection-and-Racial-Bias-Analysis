```python
import pandas as pd
```

```python
import numpy as np
```

```python
aae_data = pd.read_csv('/content/drive/My Drive/fdcl18_aae.csv', header=None, sep=',|\s+', engine='python', usecols=[0])# Use header=None and set the separator to tab
```

```python
aae_data.rename(columns={0: 'AAE_prob'}, inplace=True)
```

```python
# Concatenate the first column(AAE_prob) from aae_data to the last column of data
result = pd.concat([data, aae_data['AAE_prob']], axis=1)
```

```python
result
```

```python
from scipy.stats import pearsonr
from nltk.tokenize import TweetTokenizer
```

```python
#result1= result.head(170)
result['AAE_prob'].fillna(0, inplace=True)
```

```python
labels = ['normal', 'abusive', 'hateful', 'spam']
correlations = {}

for label in labels:
    # Create a binary column indicating the presence of the current label
    result[f'{label}_binary'] = np.where(result['Label'] == label, 1, 0)
    
    # Calculate the correlation between AAE probability and the binary column for the current label
    corr, _ = pearsonr(result['AAE_prob'], result[f'{label}_binary'])
    correlations[label] = corr

# Print the AAE correlations
print("AAE correlations for each label in the dataset:")
for label, corr in correlations.items():
    print(f"{label}: {corr:.3f}")
```

```python
# Used DWMW17 from https://github.com/t-davidson/hate-speech-and-offensive-language/tree/master/data
data_path = 'https://raw.githubusercontent.com/t-davidson/hate-speech-and-offensive-language/master/data/labeled_data.csv'
DWMW17data = pd.read_csv(data_path)
```

```python
aae_data17 = pd.read_csv('/content/drive/My Drive/aae.txt', header=None, sep=',|\s+', engine='python', usecols=[0])# Use header=None and set the separator to tab
```

```python
aae_data17.rename(columns={0: 'AAE_prob'}, inplace=True)
```

```python
aae_data17
```

```python
# Concatenate the first column(AAE_prob) from aae_data to the last column of data
result17 = pd.concat([DWMW17data, aae_data17['AAE_prob']], axis=1)
```

```python
result17
```

```python
from scipy.stats import pearsonr
result17['AAE_prob'].fillna(0, inplace=True)

##calculates the Pearson correlation coefficient between "AAE_prob" and each label in the list of labels.
## The results are stored in a dictionary called "correlations", where the keys are the labels and the values are the correlation coefficients.
labels = ['hate_speech', 'offensive_language', 'neither']
correlations = {label: pearsonr(result17['AAE_prob'], result17[label])[0] for label in labels}

print("AAE correlations for each category in the dataset:")
for label, corr in correlations.items():
    print(f"{label}: {corr:.3f}")
```
